# Практическая работа 5. Объектно-ориентированное программирование

## Задание

Класс «Цилиндр». Реализовать ввод и вывод полей данных,
вычисление объема, площади поверхности, а также вывод
информации об объекте.

## Решение

```python
class Cyl:
  # Объявляем конструктор класса
  def __init__(self, pa, pb):
    self.r=pa
    self.h=pb

  # Объявляем метод для вычисления площади поверхности фигуры
  def GetS(self):
    return 6.28 * self.r * self.r + 6.28 * self.r * self.h

  # Объявляем метод для вычисления объема
  def GetV(self):
    return 3.14*self.r*self.r*self.h

  # Объявление метода для вывода информации о фигуре
  def Info(self):
    print('*'*20)
    print('Цилиндр: \n')
    print('Размеры:\n радиус основания: %.2f x высота: %.2f' % (self.r, self.h))
    print('Площадь: %.3f' % self.GetS())
    print('Объем: %.3f' % self.GetV())
    print('*'*20)
```

В данном листинге производится объявление типа данных `Cyl` с использованием зарезервированного слова `class`. Внутри класса объявляются функции (методы класса) и переменные (поля данных класса). Следует обратить внимание на использование зарезервированного слова `self`.
Также внутри класса объявлен метод `__init__()`. Данный метод имеет специальное назначение - он вызывается в момент создания переменной данного класса автоматически. Данный метод называется конструктор класса.

Использование разработанного типа данных `Cyl`

```python
# Создание объекта p1 типа Par (используется конструктор)
c1 = Cyl(10, 20)
# Вызов метода класса
c1.Info()
```
